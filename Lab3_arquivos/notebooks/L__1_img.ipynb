# Código 1: Transformar a foto em cinza

#### Realizar importações

```python
import numpy as np
import cv2 as cv
```

#### Código principal:

```python
# Realizar a leitura da imagem
img = cv.imread('messi5.jpg',0)
cv.imshow('image',img)
k = cv.waitKey(0)
```

```python
# Ao apertar a tecla ESC, a janela é fechada
if k == 27:         # wait for ESC key to exit
    cv.destroyAllWindows()

# Ao apertar a letra 's', a imagem é salva e a janela é fechada
elif k == ord('s'): # wait for 's' key to save and exit
    cv.imwrite('messigray.png',img)
    cv.destroyAllWindows()
```
