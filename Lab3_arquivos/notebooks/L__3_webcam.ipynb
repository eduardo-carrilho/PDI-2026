# Código 2: Mudar o frame rate do vídeo

#### Realizar importações

```python
import numpy as np
import cv2 as cv
```

#### Código principal:

```python
# Iniciando a captura de vídeo
cap = cv.VideoCapture(0)
```

```python
# Se a câmera não for aberta corretamente, exibe uma mensagem de erro e encerra o programa
if not cap.isOpened():
    print("Cannot open camera")
    exit()

# Se a câmera for aberta corretamente, inicia a captura de vídeo
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    # Display the resulting frame
    cv.imshow('frame', frame)
    cv.imwrite("../parte2.jpg", frame)
    
    # Quando o usuário apertar a letra 'q', a janela é fechada
    if cv.waitKey(1) == ord('q'):
        break

    # Quando o usuário apertar a letra 'x', o vídeo é salvo e a janela é fechada
    if cv.waitKey(1) == ord('x'):
        break
```

```python
# O vídeo é liberado e as janelas são fechadas
cap.release()
cv.destroyAllWindows()
```
