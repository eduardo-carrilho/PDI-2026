```python
import numpy as np
import cv2 as cv
KERNEL_LENGTHS = [3,5,11,29]

img = cv.imread('foto-geral.jpg',1)
cv.imshow('image',img)
k = cv.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    for j in KERNEL_LENGTHS:
        for i in range(1, j, 2):
            dst = cv.blur(img, (i, i))
        cv.imwrite(f"blur_kernel{j}x{j}",dst)
        cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
KERNEL_LENGTHS = [3,5,11,29]

img = cv.imread('foto-geral.jpg',1)

for j in KERNEL_LENGTHS:
    for i in range(1, j, 2):
        dst = cv.blur(img, (i, i))
    file_name = f"foto_geral_blur_kernel{j}x{j}.jpeg"
    cv.imwrite(file_name, dst)

for j in KERNEL_LENGTHS:
    for i in range(1, j, 2):
        dst = cv.GaussianBlur(img, (i, i), 0)
    file_name = f"foto_geral_gaussian_blur_kernel{j}x{j}.jpeg"
    cv.imwrite(file_name, dst)

for j in KERNEL_LENGTHS:
    for i in range(1, j, 2):
        dst = cv.medianBlur(img, i)
    file_name = f"foto_geral_median_blur_kernel{j}x{j}.jpeg"
    cv.imwrite(file_name, dst)

for j in KERNEL_LENGTHS:
    for i in range(1, j, 2):
        dst = cv.bilateralFilter(img, i, i * 2, i / 2)
    file_name = f"foto_geral_bilateral_filter_kernel{j}x{j}.jpeg"
    cv.imwrite(file_name, dst)
```
