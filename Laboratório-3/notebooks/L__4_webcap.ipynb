# Código 2: Mudar o frame rate do vídeo

#### Realizar importações

```python
import numpy as np
import cv2 as cv
```

#### Código principal:

```python
# Iniciando a captura de vídeo
cap = cv.VideoCapture(0)
```

```python
# Obtém a largura atual do frame
width = cap.get(cv.CAP_PROP_FRAME_WIDTH)   # float

# Obtém a altura atual do frame
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT) # float

# Define a taxa de frames do vídeo em fps
fps = 30.0

# Define o codec e cria o objeto VideoWriter
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('saida.avi', fourcc, fps, (int(width),int(height)) )

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    #frame = cv.flip(frame, 0)
    # escreve o frame modificado
    out.write(frame)
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break
```

```python
# Libera tudo se o trabalho foi concluído
cap.release()
out.release()
cv.destroyAllWindows()
```
