# Código 2: Mudar o frame rate do vídeo

#### Realizar importações

```python
import cv2 as cv
import time
```

#### Código principal:

```python
# Abrindo o vídeo
cap = cv.VideoCapture('../big_buck_bunny.mp4')
```

```python
# Enquanto o vídeo estiver aberto, leia os frames e mostre-os com um delay definido para controlar a taxa de quadros (fps)
while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret==True:
        # show the frame
        cv.imshow('frame',frame)

        # Controla a taxa de quadros (fps) do vídeo, ajustando o tempo de espera entre os frames.
        #wait next frame by 40ms - 25fps
        time.sleep(1/50.0) 
        
        # Ao apertar a letra 'q', o vídeo é interrompido e a janela é fechada
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:

        break
```

```python
# Libere o vídeo e feche as janelas
cap.release()
cv.destroyAllWindows()
```
